```python
energy_files={
    'electricity':electricity,
    'chilledwater':chilledwater,
    'gas':gas,
    'hotwater':hotwater,
    'irrigation':irrigation,
    'solar':solar,
    'steam':steam,
    'water':water,
}
```

```python
# PREPROCESSING
```

```python
combined=None
```

```python
sample_buildings=None
```

```python
for energy_name, file in energy_files.items():
    df = file.copy()
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df = df.ffill().bfill()
    df_long = df.melt(
        id_vars='timestamp',
        var_name='building_id',
        value_name=energy_name
    )
    if energy_name == "electricity":
        sample_buildings = df_long['building_id'].unique()[:50]
    df_long = df_long[
        df_long['building_id'].isin(sample_buildings)
    ]
    if combined is None:
        combined = df_long
    else:
        combined = pd.merge(
            combined,
            df_long[['timestamp','building_id',energy_name]],
            on=['timestamp','building_id'],
            how='left'
        )
    combined[energy_name] = combined[energy_name].fillna(0)
```

```python
combined.columns
```

```python
combined=combined.sort_values(['building_id','timestamp'])
```

```python
# FEATURE ENGINEERING
```

```python
combined['dayofweek']=combined['timestamp'].dt.dayofweek
```

```python
combined['hour']=combined['timestamp'].dt.hour
```

```python
combined['month']=combined['timestamp'].dt.month
```

```python
combined['is_weekend']=combined['dayofweek']>5
```

```python
combined['e_rollingmean']=(
    combined.groupby('building_id')['electricity'].transform(lambda x:x.rolling(24,min_periods=1).mean())
)
```

```python
combined['elag24']=(
    combined.groupby('building_id')['electricity'].shift(24)
)
```

```python
combined['c_rollingmean']=(
    combined.groupby('building_id')['chilledwater'].transform(lambda x:x.rolling(24,min_periods=1).mean())
)
```

```python
combined['clag24']=(
    combined.groupby('building_id')['chilledwater'].shift(24)
)
```

```python
combined['eabsdeviation']=(
    combined['electricity']-combined['e_rollingmean']
).abs()
```

```python
combined['cabsdeviation']=(
    combined['chilledwater']-combined['c_rollingmean']
).abs()
```

```python
features=[
    'electricity',
    'chilledwater',
    'hotwater',
    'gas',
    'steam',
    'water',
    'solar',
    'irrigation',
    'hour','dayofweek','month','is_weekend','e_rollingmean',
    'c_rollingmean','clag24','elag24','eabsdeviation','cabsdeviation'
]
```

```python
x=combined[features].copy()
```

```python
x['is_weekend']=x['is_weekend'].astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
# MODEL TRAINING
```

```python
from sklearn.ensemble import IsolationForest
```

```python
model=IsolationForest(
    n_estimators=100,
    contamination=0.01,
    random_state=42,
    n_jobs=1
)
```

```python
model.fit(x_scaled)
```

```python
combined['anomaly']=model.predict(x_scaled)
```

```python
combined['anomaly']=combined['anomaly'].map({-1:1,1:0})
```

```python
combined['anomaly'].value_counts()
```

```python
# DATA VISUALIZATION
```

```python
sample_building = combined['building_id'].iloc[0]
temp = combined[
    combined['building_id'] == sample_building
].sort_values('timestamp')
plt.figure(figsize=(14,5))
plt.plot(
    temp['timestamp'],
    temp['electricity'],
    label='Electricity Usage',
    alpha=0.6
)
plt.scatter(
    temp[temp['anomaly']==1]['timestamp'],
    temp[temp['anomaly']==1]['electricity'],
    color='red',
    s=10,
    label='Anomaly'
)
plt.title('Electricity Consumption with Detected Anomalies')
plt.xlabel('Time')
plt.ylabel('Electricity')
plt.legend()
plt.show()
```

```python
top_buildings = (
    combined.groupby('building_id')['anomaly']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
plt.figure(figsize=(12,5))
top_buildings.plot(kind='bar')
plt.title('Top Buildings with Highest Number of Anomalies')
plt.xlabel('Building ID')
plt.ylabel('Anomaly Count')
plt.show()
```
